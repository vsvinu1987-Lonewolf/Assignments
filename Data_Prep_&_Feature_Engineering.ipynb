```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, RobustScaler
```

```python
# Initializing the dataset
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}
```

```python
df = pd.DataFrame(data)
```

**Task 1: Categorical Encoding**

```python
#Gender --> Label Encoding
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])

# Size --> Label Encoding with specific order
#Mapping ensures small = 0, medium = 1, Large = 2 as requested
size_mapping = {'Small': 0, 'Medium': 1, 'Large': 2}
df['Size'] = df['Size'].map(size_mapping)

# City --> one-hot Encoding (Drop First to avoide dummy variable trap)
df=pd.get_dummies(df, columns=['City'], drop_first=True)
```

**Task 2: Feature Scaling**

```python
# We use RobustScaler because the 'Fare' column contains an outlier (450).
# RobustScaler uses the median and IQR, making it less sensitive to extreme values
# compared to StandardScaler or Min-Max Scaler.
scaler = RobustScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
```

```python
# Displaying the final preprocessed dataframe
print("Preprocessed Dataframe:")
df
```

**Key Decisions & Justification**

**Gender:** Used LabelEncoder to convert binary text into 0 and 1.

**City:** Used get_dummies with drop_first=True. This removes one column (usually the first one alphabetically) to prevent multicollinearity, which is essential for linear models.

**Size:** A manual .map() was used instead of a generic encoder to guarantee the specific order of Small < Medium < Large.

**Scaling:** I chose RobustScaler. As we discussed in your previous question about insurance claims, StandardScaler and Min-Max would be heavily distorted by that 450 fare. RobustScaler keeps the rest of the data useful by centering it around the median.
